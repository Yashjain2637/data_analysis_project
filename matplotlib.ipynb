```python
import matplotlib.pyplot as plt
# here the pyplot is the submodule which is used to plot the graphs
```

```python
import pandas as pd
```

```python
# 1. line plot
data={
    'date':[20,21,22,23,24,25],
    'sales':[200,210,300,400,500,350],
    'stock':[250,300,350,230,400,200]
}
df=pd.DataFrame(data)
```

```python
# line plot is used when we have to visualise the time series data
# into the time series data we should always plot the time on the x - axis
# stock --> y - axis
# syntax ==> plt.plot(x-axis,y-axis)
```

```python
plt.plot(df['date'],df['sales'],label='sales')
plt.plot(df['date'],df['stock'],label='stock')
plt.title('Date vs sales Graph')   # ==> it is used to give title to the graph
plt.xlabel('Date')                 # ==> it is used to label the x - axis
plt.ylabel('Sales')                # ==> it is used to label the y - axis
plt.legend()                       # ==> it is used to show legends at the top corner
plt.show()                         # ==> it is used to show the graph which is built
```

```python
# 2. bar plot  ==> bar plot is used when we want to plot a graph between the categorical and numerical frequency of each the category
student_data={
    'student_name':['A','B','C','D','E','F','G','H'],
    'marks':[20,35,40,23,55,40,36,78]
}
sdf=pd.DataFrame(student_data)
```

```python
plt.bar(sdf['student_name'],sdf['marks'])
# plt.barh(sdf['student_name'],sdf['marks']) ==> plt.barh is used when we have more number of categories
plt.title('Student Marks')
plt.xlabel('Student Name')
plt.ylabel('Marks')
plt.show()
```

```python
# 3. scatter plot ==> it is used to find the relationship between the two features
data ={
    'area':[200,300,400,500],
    'price':[20,30,40,50],
    'price1':[50,40,30,20]
}
hdf=pd.DataFrame(data)

# area -> price -> direct relationship -> area increases, price increases
# area -> price1 -> indirect relationship -> area increases , price1 decreases
```

```python
plt.scatter(hdf['area'],hdf['price1'])
```

```python
plt.scatter(hdf['area'],hdf['price'])
```

```python
plt.scatter(hdf['area'],hdf['price1'],label='Area')
plt.scatter(hdf['area'],hdf['price'],label='Sales')
plt.xlabel('Area')
plt.ylabel('price')
plt.legend()
```
